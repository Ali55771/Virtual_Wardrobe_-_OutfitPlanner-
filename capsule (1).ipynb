```python
import itertools
import pandas as pd

# 50 Famous Colors (popular in clothing)
colors = [
    'White', 'Black', 'Red', 'Blue', 'Green', 'Yellow', 'Orange', 'Pink', 'Purple', 'Brown',
    'Beige', 'Grey', 'Navy Blue', 'Light Blue', 'Dark Blue', 'Light Green', 'Dark Green', 'Burgundy',
    'Cream', 'Peach', 'Tan', 'Coral', 'Gold', 'Silver', 'Emerald Green'
]

# Function to determine if two colors are complementary (based on color theory)
def is_complementary(color1, color2):
    complementary_pairs = {
        ('Red', 'Green'), ('Blue', 'Orange'), ('Yellow', 'Purple'),
        ('Pink', 'Green'), ('Red', 'Cyan'), ('Blue', 'Yellow')
    }
    return (color1, color2) in complementary_pairs or (color2, color1) in complementary_pairs

# Function to determine if two colors are analogous (next to each other on the color wheel)
def is_analogous(color1, color2):
    analogous_pairs = {
        ('Red', 'Orange'), ('Red', 'Pink'), ('Blue', 'Light Blue'),
        ('Green', 'Light Green'), ('Green', 'Olive'), ('Blue', 'Turquoise')
    }
    return (color1, color2) in analogous_pairs or (color2, color1) in analogous_pairs

# Function to determine if two colors are monochromatic (same color, different shades)
def is_monochromatic(color1, color2):
    return color1 == color2  # Monochromatic is just the same color

# Function to calculate the compatibility score based on the rules
def get_color_score(shirt_color, pant_color, shoe_color, jacket_color):
    score = 0

    # Check shirt-pant compatibility
    if is_complementary(shirt_color, pant_color):
        score += 3
    elif is_analogous(shirt_color, pant_color):
        score += 2
    elif is_monochromatic(shirt_color, pant_color):
        score += 1

    # Check shirt-shoe compatibility
    if is_complementary(shirt_color, shoe_color):
        score += 3
    elif is_analogous(shirt_color, shoe_color):
        score += 2
    elif is_monochromatic(shirt_color, shoe_color):
        score += 1

    # Check shirt-jacket compatibility
    if is_complementary(shirt_color, jacket_color):
        score += 3
    elif is_analogous(shirt_color, jacket_color):
        score += 2
    elif is_monochromatic(shirt_color, jacket_color):
        score += 1

    return score

# Generate all possible combinations of colors for shirts, pants, shoes, jackets
combinations = list(itertools.product(colors, repeat=4))  # 4 items: Shirt, Pant, Shoe, Jacket

# List to hold valid combinations with their scores
valid_combinations = []

# Generate all valid combinations with their score
for combo in combinations:
    shirt_color, pant_color, shoe_color, jacket_color = combo
    score = get_color_score(shirt_color, pant_color, shoe_color, jacket_color)

    # Only include combinations that have a non-zero score
    if score > 0:
        valid_combinations.append([shirt_color, pant_color, shoe_color, jacket_color, score])

# Create a DataFrame for storing the valid combinations
df = pd.DataFrame(valid_combinations, columns=['Shirt Color', 'Pant Color', 'Shoe Color', 'Jacket Color', 'Score'])

# Sort by the score (best combinations first)
df_sorted = df.sort_values(by='Score', ascending=False)

# Save to CSV
df_sorted.to_csv('best_color_combinations.csv', index=False)

# Output the number of valid combinations
print(f"Total valid color combinations: {len(valid_combinations)}")
```

```python
import pandas as pd
from collections import defaultdict

# Hardcoded selections with psychologically optimized colors
HARDCODED_SELECTIONS = {
    'shirts': ['Light Blue', 'White', 'Light Pink', 'Light Yellow'],  # Calming and professional
    'pants': ['Dark Blue', 'Charcoal', 'Dark Gray'],  # Professional and slimming
    'shoes': ['Black', 'Dark Brown'],  # Neutral and versatile
    'jackets': ['Navy', 'Light Gray']  # Complementary to shirts
}

# Color psychology attributes (Energy, Calmness, Professionalism)
COLOR_PSYCHOLOGY = {
    # Blues (calming, trustworthy)
    'Blue': (0.6, 0.8, 0.9),
    'Light Blue': (0.5, 0.9, 0.8),  # Very calming
    'Dark Blue': (0.7, 0.7, 1.0),  # Most professional
    'Navy': (0.7, 0.7, 1.0),

    # Whites/Grays (neutral, professional)
    'White': (0.5, 0.9, 0.9),
    'Light Gray': (0.4, 0.9, 0.8),
    'Dark Gray': (0.3, 0.7, 1.0),
    'Charcoal': (0.3, 0.7, 1.0),

    # Pinks (approachable, friendly)
    'Pink': (0.7, 0.7, 0.6),
    'Light Pink': (0.6, 0.8, 0.5),  # Soft and calming

    # Yellows (energizing but less professional)
    'Yellow': (0.9, 0.5, 0.5),
    'Light Yellow': (0.8, 0.7, 0.6),  # Softer version

    # Browns (earthy, reliable)
    'Dark Brown': (0.4, 0.8, 0.8),
    'Black': (0.2, 0.7, 1.0)  # Most formal
}

def get_psychology(color):
    """Get psychological attributes for a color"""
    base_color = color.split(' ')[-1]  # Handle 'Light/Dark' prefixes
    return COLOR_PSYCHOLOGY.get(base_color, (0.5, 0.5, 0.7))

def color_harmony(color1, color2):
    """Calculate harmony score between two colors based on color theory"""
    # Complementary pairs (high contrast = high harmony)
    complements = {
        'Blue': 'Orange',
        'Red': 'Green',
        'Yellow': 'Purple'
    }

    base1 = color1.split(' ')[-1]
    base2 = color2.split(' ')[-1]

    # Check complements
    if base1 in complements and complements[base1] == base2:
        return 2.5
    if base2 in complements and complements[base2] == base1:
        return 2.5

    # Analogous colors (similar hues)
    analogous_groups = [
        ['Blue', 'Light Blue', 'Dark Blue', 'Navy'],
        ['Gray', 'Light Gray', 'Dark Gray', 'Charcoal'],
        ['Brown', 'Dark Brown', 'Black']
    ]

    for group in analogous_groups:
        if base1 in group and base2 in group:
            return 2.0  # Harmonious similar colors

    # Light-Dark contrast
    lightness1 = 0 if 'Light' in color1 else (1 if 'Dark' in color1 else 0.5)
    lightness2 = 0 if 'Light' in color2 else (1 if 'Dark' in color2 else 0.5)
    if abs(lightness1 - lightness2) > 0.6:
        return 1.8  # Good contrast

    return 1.0  # Neutral combination

def calculate_outfit_score(shirt, pants, shoes, jacket):
    """Calculate overall outfit score considering color harmony and psychology"""
    # Weighted color harmony
    harmony = (
        3.0 * color_harmony(shirt, pants) +  # Most important
        2.0 * color_harmony(pants, shoes) +
        2.0 * color_harmony(shirt, jacket) +
        1.5 * color_harmony(jacket, pants)
    )

    # Psychological impact
    shirt_e, shirt_c, shirt_p = get_psychology(shirt)
    pants_e, pants_c, pants_p = get_psychology(pants)
    shoes_e, shoes_c, shoes_p = get_psychology(shoes)
    jacket_e, jacket_c, jacket_p = get_psychology(jacket)

    # Weighted averages
    energy = 0.4*shirt_e + 0.2*pants_e + 0.1*shoes_e + 0.3*jacket_e
    calm = 0.4*shirt_c + 0.2*pants_c + 0.1*shoes_c + 0.3*jacket_c
    professional = 0.4*shirt_p + 0.2*pants_p + 0.1*shoes_p + 0.3*jacket_p

    # Ideal professional balance targets
    mood_balance = 2.5 - abs(energy - 0.5) - abs(calm - 0.7)
    prof_score = professional * 1.5

    # Final score
    final_score = harmony * mood_balance * prof_score
    return (
        round(max(0, final_score), 2),
        round(energy, 2),
        round(calm, 2),
        round(professional, 2)
    )

def generate_outfits_from_selections(shirts, pants, shoes, jackets, num_outfits=4):
    """Generate top outfits from selected items"""
    used_colors = defaultdict(set)
    outfits = []

    # Generate all possible combinations
    candidates = []
    for shirt in shirts:
        for pant in pants:
            for shoe in shoes:
                for jacket in jackets:
                    # Skip clashing combinations
                    if pant == jacket or shoe == jacket:
                        continue

                    score, energy, calm, prof = calculate_outfit_score(shirt, pant, shoe, jacket)
                    candidates.append((score, shirt, pant, shoe, jacket, energy, calm, prof))

    # Sort by best scores first
    candidates.sort(reverse=True, key=lambda x: x[0])

    # Select top unique outfits
    for candidate in candidates:
        score, shirt, pant, shoe, jacket, energy, calm, prof = candidate

        colors = {shirt, pant, shoe, jacket}
        if (not colors.intersection(used_colors['shirt']) and
            not colors.intersection(used_colors['pants']) and
            not colors.intersection(used_colors['shoes']) and
            not colors.intersection(used_colors['jacket'])):

            outfits.append({
                'shirt': shirt,
                'pants': pant,
                'shoes': shoe,
                'jacket': jacket,
                'score': score,
                'energy': energy,
                'calm': calm,
                'professional': prof
            })

            # Track used colors
            used_colors['shirt'].add(shirt)
            used_colors['pants'].add(pant)
            used_colors['shoes'].add(shoe)
            used_colors['jacket'].add(jacket)

            if len(outfits) >= num_outfits:
                break

    return pd.DataFrame(outfits)

if __name__ == "__main__":
    # Use hardcoded psychologically optimized selections
    shirts = HARDCODED_SELECTIONS['shirts']
    pants = HARDCODED_SELECTIONS['pants']
    shoes = HARDCODED_SELECTIONS['shoes']
    jackets = HARDCODED_SELECTIONS['jackets']

    print("Using psychologically optimized selections:")
    print(f"Shirts: {', '.join(shirts)}")
    print(f"Pants: {', '.join(pants)}")
    print(f"Shoes: {', '.join(shoes)}")
    print(f"Jackets: {', '.join(jackets)}\n")

    # Generate and display outfits
    outfits = generate_outfits_from_selections(shirts, pants, shoes, jackets)
    print("Top 4 Psychologically-Optimized Outfit Combinations")
    print(outfits[['shirt', 'pants', 'shoes', 'jacket', 'score', 'energy', 'calm', 'professional']].to_string(index=False))
```

```python
import pandas as pd
import random
from collections import defaultdict

# Complete color palette with psychological attributes
COLOR_PALETTE = {
    'shirt': [
        'Pink', 'Light Pink', 'Dark Pink', 'Red', 'Light Red', 'Dark Red',
        'Blue', 'Light Blue', 'Dark Blue', 'Green', 'Light Green', 'Dark Green',
        'Yellow', 'Light Yellow', 'Dark Yellow', 'Orange', 'Light Orange', 'Dark Orange',
        'Purple', 'Light Purple', 'Dark Purple', 'White', 'Gray', 'Light Gray', 'Dark Gray',
        'Black', 'Beige', 'Light Beige', 'Dark Beige'
    ],
    'pants': [
        'Black', 'Dark Gray', 'Gray', 'Light Gray', 'White', 'Dark Blue', 'Blue',
        'Dark Brown', 'Brown', 'Light Brown', 'Beige', 'Dark Beige', 'Khaki',
        'Dark Green', 'Olive', 'Charcoal'
    ],
    'shoes': [
        'Black', 'Dark Brown', 'Brown', 'Tan', 'White', 'Gray', 'Dark Gray',
        'Oxblood', 'Burgundy', 'Dark Blue', 'Dark Green'
    ],
    'jacket': [
        'Black', 'Navy', 'Dark Gray', 'Charcoal', 'Brown', 'Dark Brown', 'Beige',
        'Light Gray', 'White', 'Olive', 'Dark Green', 'Camel', 'Tan'
    ]
}

# Psychological attributes (Energy, Calmness, Professionalism)
COLOR_PSYCHOLOGY = {
    'Pink': (0.7, 0.7, 0.6),
    'Light Pink': (0.6, 0.8, 0.5),
    'Dark Pink': (0.7, 0.6, 0.7),
    'Red': (0.9, 0.3, 0.7),
    'Light Red': (0.8, 0.5, 0.6),
    'Dark Red': (0.7, 0.6, 0.8),
    'Blue': (0.6, 0.8, 0.9),
    'Light Blue': (0.5, 0.9, 0.8),
    'Dark Blue': (0.7, 0.7, 1.0),
    'Green': (0.5, 0.8, 0.7),
    'Light Green': (0.4, 0.9, 0.6),
    'Dark Green': (0.6, 0.7, 0.8),
    'Yellow': (0.9, 0.5, 0.5),
    'Light Yellow': (0.8, 0.7, 0.6),
    'Dark Yellow': (0.7, 0.6, 0.7),
    'Orange': (0.8, 0.5, 0.6),
    'Light Orange': (0.7, 0.7, 0.5),
    'Dark Orange': (0.7, 0.6, 0.7),
    'Purple': (0.7, 0.6, 0.8),
    'Light Purple': (0.6, 0.8, 0.7),
    'Dark Purple': (0.7, 0.7, 0.9),
    'White': (0.5, 0.9, 0.9),
    'Gray': (0.3, 0.8, 0.9),
    'Light Gray': (0.4, 0.9, 0.8),
    'Dark Gray': (0.3, 0.7, 1.0),
    'Black': (0.2, 0.7, 1.0),
    'Beige': (0.4, 0.9, 0.7),
    'Light Beige': (0.3, 1.0, 0.6),
    'Dark Beige': (0.4, 0.8, 0.8)
}

def get_psychology(color):
    base_color = color.split(' ')[-1]  # Handle 'Light/Dark' prefixes
    return COLOR_PSYCHOLOGY.get(base_color, (0.5, 0.5, 0.7))

def color_harmony(color1, color2):
    # Complementary colors
    complements = {
        'Red': 'Green',
        'Blue': 'Orange',
        'Yellow': 'Purple',
        'Green': 'Red',
        'Orange': 'Blue',
        'Purple': 'Yellow'
    }

    base1 = color1.split(' ')[-1]
    base2 = color2.split(' ')[-1]

    # Check complements
    if base1 in complements and complements[base1] == base2:
        return 2.5
    if base2 in complements and complements[base2] == base1:
        return 2.5

    # Analogous colors (same color family)
    color_families = {
        'Red': ['Pink', 'Red', 'Orange'],
        'Blue': ['Blue', 'Light Blue', 'Dark Blue', 'Navy'],
        'Green': ['Green', 'Light Green', 'Dark Green', 'Olive'],
        'Neutral': ['Black', 'White', 'Gray', 'Beige', 'Brown', 'Tan']
    }

    for family in color_families.values():
        if base1 in family and base2 in family:
            return 2.0 if abs(family.index(base1) - family.index(base2)) == 1 else 1.8

    # Light-Dark contrast
    lightness1 = 0 if 'Light' in color1 else (1 if 'Dark' in color1 else 0.5)
    lightness2 = 0 if 'Light' in color2 else (1 if 'Dark' in color2 else 0.5)
    if abs(lightness1 - lightness2) > 0.6:
        return 1.7

    return 1.0  # Neutral combination

def calculate_outfit_score(shirt, pants, shoes, jacket):
    # Color harmony (weighted)
    harmony = (
        3.0 * color_harmony(shirt, pants) +
        2.0 * color_harmony(pants, shoes) +
        2.0 * color_harmony(shirt, jacket) +
        1.5 * color_harmony(jacket, pants)
    )

    # Psychological balance
    shirt_e, shirt_c, shirt_p = get_psychology(shirt)
    pants_e, pants_c, pants_p = get_psychology(pants)
    shoes_e, shoes_c, shoes_p = get_psychology(shoes)
    jacket_e, jacket_c, jacket_p = get_psychology(jacket)

    # Calculate weighted scores
    energy = 0.4 * shirt_e + 0.2 * pants_e + 0.1 * shoes_e + 0.3 * jacket_e
    calm = 0.4 * shirt_c + 0.2 * pants_c + 0.1 * shoes_c + 0.3 * jacket_c
    professional = 0.4 * shirt_p + 0.2 * pants_p + 0.1 * shoes_p + 0.3 * jacket_p

    # Penalize too many same-hue colors
    hue_counts = defaultdict(int)
    for color in [shirt, pants, shoes, jacket]:
        base_color = color.split(' ')[-1]
        hue_counts[base_color] += 1
    hue_penalty = sum(max(0, count-1) * 0.3 for count in hue_counts.values())

    # Calculate final score
    mood_balance = 2.5 - abs(energy - 0.6) - abs(calm - 0.7)
    prof_score = professional * 1.5
    final_score = harmony * mood_balance * prof_score - hue_penalty

    # Ensure the final score and other metrics are properly returned
    return (round(max(0, final_score), 2), round(energy, 2), round(calm, 2), round(professional, 2))


def generate_outfits(num_outfits=4):
    used_colors = defaultdict(set)
    outfits = []

    # Generate candidates
    candidates = []
    for _ in range(3000):
        shirt = random.choice(COLOR_PALETTE['shirt'])
        pants = random.choice(COLOR_PALETTE['pants'])
        shoes = random.choice(COLOR_PALETTE['shoes'])
        jacket = random.choice(COLOR_PALETTE['jacket'])

        # Skip invalid combinations
        if pants == jacket or shoes == jacket:
            continue

        score, energy, calm, prof = calculate_outfit_score(shirt, pants, shoes, jacket)
        candidates.append((score, shirt, pants, shoes, jacket, energy, calm, prof))

    # Sort and select unique outfits
    candidates.sort(reverse=True, key=lambda x: x[0])

    for candidate in candidates:
        score, shirt, pants, shoes, jacket, energy, calm, prof = candidate

        # Check color uniqueness
        colors = {shirt, pants, shoes, jacket}
        if (not colors.intersection(used_colors['shirt']) and
            not colors.intersection(used_colors['pants']) and
            not colors.intersection(used_colors['shoes']) and
            not colors.intersection(used_colors['jacket'])):

            outfits.append({
                'shirt': shirt,
                'pants': pants,
                'shoes': shoes,
                'jacket': jacket,
                'score': score,
                'energy': energy,
                'calm': calm,
                'professional': prof
            })

            # Mark colors as used
            used_colors['shirt'].add(shirt)
            used_colors['pants'].add(pants)
            used_colors['shoes'].add(shoes)
            used_colors['jacket'].add(jacket)

            if len(outfits) >= num_outfits:
                break

    return pd.DataFrame(outfits)

# Generate and display outfits
outfits = generate_outfits()
print("Top 4 Psychologically-Optimized Outfit Combinations")
print(outfits[['shirt', 'pants', 'shoes', 'jacket', 'score', 'energy', 'calm', 'professional']].to_string(index=False))
```
